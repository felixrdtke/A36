```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load the Data ---
# delim_whitespace=True is crucial here because ngspice uses spaces, not commas!
file_path = 'transistor0.3_0.5.csv'
try:
    df = pd.read_csv(file_path, delim_whitespace=True)
except FileNotFoundError:
    print(f"Error: Could not find '{file_path}'. Make sure it is in the same folder as this script!")
    exit()

# Print the columns so you can verify what ngspice named the headers
print("Successfully loaded data. Detected columns:", df.columns.tolist())

# --- 2. Extract Columns Dynamically ---
# Using .iloc ensures the script works perfectly even if ngspice 
# named the headers something weird like "dc1.@m.xm1.m0[id]"
v_gate = df.iloc[:, 0]    # 1st Column: Swept Gate Voltage
id_curve = df.iloc[:, 1]  # 2nd Column: Drain Current (Id)
gds_curve = df.iloc[:, 2] # 3rd Column: Output Conductance (gds)
gm_curve = df.iloc[:, 3]  # 4th Column: Transconductance (gm)


# --- 3. Style and Plot Setup ---
# Applying the official RWTH Aachen color palette
rwth_blue = '#00549F'
rwth_red = '#CC071E'
rwth_green = '#57AB27'

# Create a 2-row, 1-column layout that shares the X-axis
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(8, 8), sharex=True)

# --- Top Plot: Drain Current (Id) ---
ax1.plot(v_gate, id_curve, color=rwth_blue, linewidth=2.5, label='Drain Current ($I_D$)')
ax1.set_ylabel('Current (Amps)', fontsize=12, fontweight='bold')
ax1.set_title('Transistor DC Sweep Characteristics', fontsize=14, fontweight='bold')
ax1.grid(True, which='both', linestyle='--', alpha=0.6)
ax1.legend(loc='upper left', fontsize=11)
ax1.tick_params(axis='y', labelsize=11)

# --- Bottom Plot: Conductances (gds) ---
ax2.plot(v_gate, gds_curve, color=rwth_green, linewidth=2.5, label='Output Conductance ($g_{ds}$)')
ax2.set_xlabel('Gate Voltage (V)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Siemens', fontsize=12, fontweight='bold')
ax2.grid(True, which='both', linestyle='--', alpha=0.6)
ax2.legend(loc='upper left', fontsize=11)
ax2.tick_params(axis='both', labelsize=11)

# --- Bottom Plot: Conductances (gm) ---
ax3.plot(v_gate, gm_curve, color=rwth_red, linewidth=2.5, label='Transconductance ($g_m$)')
ax3.set_xlabel('Gate Voltage (V)', fontsize=12, fontweight='bold')
ax3.set_ylabel('Conductance (Siemens)', fontsize=12, fontweight='bold')
ax3.grid(True, which='both', linestyle='--', alpha=0.6)
ax3.legend(loc='upper left', fontsize=11)
ax3.tick_params(axis='both', labelsize=11)

# --- Final Render ---
# tight_layout prevents the labels from overlapping
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. User Configuration ---
files_to_plot = [
    {"path": "transistor0.3_0.5.csv", "label": "Transistor 1 (W/L=0.5/0.3)", "color": "#00549F"}, # RWTH Blue
    {"path": "transistor0.3_1.csv",     "label": "Transistor 2 (W/L=1/0.3)",              "color": "#CC071E"}, # RWTH Red
    {"path": "transistor0.4_0.5.csv",     "label": "Transistor 3 (W/L=0.5/0.4)",              "color": "#57AB27"}, # RWTH Green
    {"path": "transistor0.4_1.csv",     "label": "Transistor 4 (W/L=1/0.3)",              "color": "#E30066"}  # RWTH Magenta
]


# --- 2. Styling (RWTH Style configuration) ---
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica'] 
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10

# Create 3 stacked subplots (Removed sharex=True so each gets its own X-axis)
fig, (ax_id, ax_gm, ax_gds) = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

# --- 3. Data Loading and Plotting ---
for item in files_to_plot:
    try:
        df = pd.read_csv(item["path"], delim_whitespace=True, header=None)
        
        v_gate    = df.iloc[:, 0]  # Column 0: Swept Gate Voltage
        id_curve  = df.iloc[:, 1]  # Column 1: Drain Current (Id)
        gds_curve = df.iloc[:, 3]  # Column 3: Output Conductance (gds)
        gm_curve  = df.iloc[:, 5]  # Column 5: Transconductance (gm)
        
        # Plot 1
        ax_id.plot(v_gate, id_curve, color=item["color"], linewidth=2, label=item["label"])
        # Plot 2
        ax_gm.plot(v_gate, gm_curve, color=item["color"], linewidth=2, label=item["label"])
        # Plot 3
        ax_gds.plot(v_gate, gds_curve, color=item["color"], linewidth=2, label=item["label"])
        
    except FileNotFoundError:
        print(f"Warning: Could not find '{item['path']}'. Skipping this file.")

# --- 4. Subplot Formatting ---

# Top Plot (Id)
ax_id.set_title('Transistor Comparison: Drain Current ($I_D$)', fontweight='bold')
ax_id.set_ylabel('$I_D$ (A)', fontweight='bold')
ax_id.set_xlabel('Gate Voltage $V_{GS}$ (V)', fontweight='bold') # Added X-axis label
#ax_id.set_yscale('log') 
ax_id.grid(True, which='both', linestyle='--', alpha=0.5)
ax_id.legend(loc='best')

# Middle Plot (gm)
ax_gm.set_title('Transistor Comparison: Transconductance ($g_m$)', fontweight='bold')
ax_gm.set_ylabel('$g_m$ (S)', fontweight='bold')
ax_gm.set_xlabel('Gate Voltage $V_{GS}$ (V)', fontweight='bold') # Added X-axis label
ax_gm.grid(True, which='both', linestyle='--', alpha=0.5)
ax_gm.legend(loc='best')

# Bottom Plot (gds)
ax_gds.set_title('Transistor Comparison: Output Conductance ($g_{ds}$)', fontweight='bold')
ax_gds.set_ylabel('$g_{ds}$ (S)', fontweight='bold')
ax_gds.set_xlabel('Gate Voltage $V_{GS}$ (V)', fontweight='bold')
ax_gds.grid(True, which='both', linestyle='--', alpha=0.5)
ax_gds.legend(loc='best')

# Adjust layout so labels don't overlap and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load the Data ---
file_path = 'transistor0.3_1.csv'

try:
    # delim_whitespace=True handles the space separation from ngspice
    df = pd.read_csv(file_path, delim_whitespace=True, header=None)
except FileNotFoundError:
    print(f"Error: Could not find '{file_path}'. Make sure it's in the correct directory.")
    exit()

# Extract variables based on the 10-column data structure
v_gate = df.iloc[:, 0]  # Column 0: Swept Gate Voltage (VGS)
c_gs   = df.iloc[:, 1]  # Column 1: Gate-Source Capacitance (Cgs)
c_gd   = df.iloc[:, 3]  # Column 3: Gate-Drain Capacitance (Cgd)

# Convert from Farads to Femtofarads (fF) for cleaner Y-axis readable units
c_gs_fF = c_gs * 1e15
c_gd_fF = c_gd * 1e15

# --- 2. Styling (RWTH Corporate Configurations) ---
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica']
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

rwth_blue = '#00549F'
rwth_red  = '#CC071E'

# Create the figure window
fig, ax = plt.subplots(figsize=(8, 6))

# --- 3. Plotting ---
ax.plot(v_gate, c_gs_fF, color=rwth_blue, linewidth=2.5, label='Gate-Source Cap ($C_{gs}$)')
ax.plot(v_gate, c_gd_fF, color=rwth_red,  linewidth=2.5, label='Gate-Drain Cap ($C_{gd}$)')

# --- 4. Formatting ---
ax.set_title('Transistor Intrinsic Capacitances vs. $V_{GS}$', fontweight='bold', pad=15)
ax.set_xlabel('Gate Voltage $V_{GS}$ (V)', fontweight='bold')
ax.set_ylabel('Capacitance (fF)', fontweight='bold')

# Enable structured background grids
ax.grid(True, which='both', linestyle='--', alpha=0.5)

# Add a clean legend box
ax.legend(loc='best', frameon=True, facecolor='white', edgecolor='none')

# Adjust layout boundaries tightly
plt.tight_layout()
plt.show()
```
